```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create sample data
np.random.seed(42)
data = {
    'Month': ['Jan', 'Fev', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': np.random.randint(1000, 5000, 12),
    'Expenses': np.random.randint(500, 3000, 12),
    'Profit': np.random.randint(200, 2000, 12),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 12),
}

df = pd.DataFrame(data)
df['Profit_Margin'] = (df['Profit'] / df['Sales'] * 100).round(2)

# print first rows
df.head()
```

```python
# create multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# plot 1: sales by month
axes[0, 0].bar(df['Month'], df['Sales'], color='skyblue', alpha=0.7)
axes[0, 0].set_title('Monthly Sales')
axes[0, 0].set_ylabel('Sales Amount')

# plot 2: sales vs expenses scatter
axes[0, 1].scatter(df['Sales'], df['Expenses'], color='coral', alpha=0.6)
axes[0, 1].set_title('Sales vs Expenses')
axes[0, 1].set_xlabel('Sales')
axes[0, 1].set_ylabel('Expenses')

# plot 3: profit distribution
axes[1, 0].hist(df['Profit'], bins=8, edgecolor='black', alpha=0.7, color='lightgreen')
axes[1, 0].set_title('Profit Distribution')
axes[1, 0].set_xlabel('Profit')
axes[1, 0].set_ylabel('Frequency')

# plot 3: sales by region
region_sales = df.groupby('Region')['Sales'].sum()
axes[1, 1].pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('Sales by region')

plt.tight_layout()
plt.show()
```
